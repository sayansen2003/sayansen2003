```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df.describe(include='all')
```

```python
# Checking for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]

# Handling missing values
# For simplicity, we will drop rows with missing values in 'embarked'
df.dropna(subset=['embarked'], inplace=True)

# Fill missing values in 'age' with the median age
df['age'].fillna(df['age'].median(), inplace=True)

# Drop the 'deck' column due to a large number of missing values
df.drop(columns=['deck'], inplace=True)

# Drop the 'embark_town' column as 'embarked' is more useful
df.drop(columns=['embark_town'], inplace=True)

# Drop 'alive' column as it is redundant with 'survived'
df.drop(columns=['alive'], inplace=True)
```

```python
# One-hot encode categorical variables
df = pd.get_dummies(df, columns=['sex', 'embarked', 'class', 'who', 'adult_male', 'alone'], drop_first=True)

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Normalize numerical columns: 'age', 'fare'
df[['age', 'fare']] = scaler.fit_transform(df[['age', 'fare']])

df.head()
```

```python
# Feature: Family size
df['family_size'] = df['sibsp'] + df['parch'] + 1

# Drop 'sibsp' and 'parch' as they are now redundant
df.drop(columns=['sibsp', 'parch'], inplace=True)

df.head()
```
